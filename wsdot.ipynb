```python
from os import listdir
from os.path import isfile, join
import json
import pandas as pd
path='data/'
df=pd.DataFrame()

onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
for filename in onlyfiles:
    with open('data/'+filename) as json_data:
        try:
            print(filename)
            data=json.load(json_data)
            
            tempdf=pd.DataFrame(data['vessellist'])        
            df=df.append(tempdf)
        except Exception:
            pass
            
df.to_csv('filtered_data.csv')        
#df=df[(df.vesselID==20)]
print(df['inservice'])
```

```python
from datetime import timedelta
import pandas as pd
import datetime

dff=pd.read_csv('filtered_dataID.csv')
#dff=dff[(dff.vesselID==20)]
#print(dff['datetime'])
dff2=dff.groupby(['leftdockdt','vesselID'])
dff2=dff2.agg(lambda x: x.iloc[-1])
dff2['arrival_time'] = dff2['datetime'].apply(lambda x: x.split(' ')[3])

dff2['arrival_time']=dff2['arrival_time'].apply(lambda x: datetime.datetime.strptime(x, '%H:%M'))

dff2['arrival_time']=dff2['arrival_time']+timedelta(minutes=1)
dff2['arrival_time']=dff2['arrival_time'].apply(lambda x: datetime.datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S').replace(year=2015).time())
dff2['temp'] = dff2['temp'].apply(lambda x: x.split(' ')[1])
dff2['temp']=dff2['temp'].str[:-2] + ' ' + dff2['temp'].str[-2:]
dff2['temp']=dff2['temp'].apply(lambda x: datetime.datetime.strptime(x, '%I:%M %p').time())


#dff2['dura2']=dff2['temp']-dff2['arrival_time']
#print(dff2['dura'])
print(dff2)

#dff2=dff2[(dff2.vesselID==20)]
#print(dff2.index['leftdockdt'])
   
```
